### Checking for Missing Values

```python
def missingValueReport(column):
    #Function looks at a particular column within the fuel_df dataframe and counts the missing values within the column.

    missingValues = fuel_df[f'{column}'].isna().sum()
    
    report = f"Column Name: {column} \
    \nNumber of Missing Values: {missingValues} \
    \nTotal Value Count in Column: {len(fuel_df[f'{column}'])}\n"
    
    return report
```

```python
for columnName in fuel_df.columns:
    print(missingValueReport(columnName))
```

### Filling in EV_Network Missing Values

We noticed that the EV_Network and EV_Network_Web columns have a large number of missing values. Perhaps the missing values in the EV_Network are supposed to be Non-Networked, which in turn makes them not have a website.

```python
#uniqueValueReport() takes in a column name from the fuel_df dataframe and returns the value counts for the unique values within that column**

def uniqueValueReport(columnName):
    return fuel_df[columnName].groupby(fuel_df[columnName]).count()
```

**Checking to see the different EV Network names and their counts**

```python
uniqueValueReport("EV_Network")
```

**Checking the missing EV_Network values by fuel type**  
It appears the Electric Vehicle Charging Stations are only missing 6 EV_Network Values. There is a chance that they have a EV Network if they have a value in the EV_Network_Web column. If not they would be placed in the Non-Networked category for EV_Network

```python
for fuel_type in fuel_df.Fuel_Type_Code.unique():
    print(f"Fuel Type: {fuel_type}")
    print("Total Missing EV_Network Values: " + str(len(fuel_df.query(f"Fuel_Type_Code == '{fuel_type}' and EV_Network.isna()"))))
    print("Total Values by Fuel Type: " + str(len(fuel_df.query(f"Fuel_Type_Code == '{fuel_type}'"))) + "\n")
```

**Checking to see whether or not the current Non-Networked alternative fuel stations have an EV_Network_Web value associated with them. 
Turns out all of the Non-Networked Alternative Fuel Stations do not have a EV_Network_Web Value**  
`Missing_EV_Network_Web_Count` - Gets the missing value count of the EV_Network_Web based on the Non-Networked values in the EV_Network Column  
`Proportion` - Gets the proportion of the missing EV_Network_Web_Count in relation to the number of Non-Networked EV_Network websites

```python
Missing_EV_Network_Web_Count = fuel_df.query("EV_Network == 'Non-Networked'").EV_Network_Web.isna().count()
Proportion = Missing_EV_Network_Web_Count / len(fuel_df.query("EV_Network == 'Non-Networked'").EV_Network_Web)
print(f"Number of Non-Networked Alternative Fuel Stations that do not have a EV_Network_Web value: {Missing_EV_Network_Web_Count}.\nPropotion of Non-Networked Alternative Fuel Stations: {Proportion}")
```

**Checking to see if the missing values within the EV_Network column have any values in the EV_Network_Web column.**

```python
print("Number of Missing EV_Network Values: ", len(fuel_df[fuel_df.EV_Network.isna()]))
print(f"Number of Missing EV_Network_Web Values for the Number of Missing EV_Network Values: {fuel_df[fuel_df.EV_Network.isna()].EV_Network_Web.isna().count()}")
```

**With:  
(1) Non-Networked EV_Network values not having a EV Network_Web value associated with them, and  
(2) The missing EV_Network values not having a EV_Network_Web value associated with them,and

### Checking for Outliers in State ZIP Codes, and Longitude-Latitude Coorinates
**With most the visualizations having to do with locational data, we wanted to make sure that there were not any outliers in tersm of ZIP codes and Longitude-Latitude points. If there is only a small number of outliers we will end up dropping them**

```python
def ZIPCodeCheck(State, valueRange):
    #This function takses in a state, and a valueRange in order to return a given number of the highest and lowest numbered ZIP Codes.
    #The state and number of values it prints out is dependent on the parameters passed in the parameter list.
    #The function prints out the state, the lower numbered and higher numbered ZIP codes and the corresponding index position.
    
    ZIPCode_mins = fuel_df.query(f"State == '{State}'").sort_values("ZIP", ascending=True).head(valueRange).ZIP
    ZIPCode_maxs = fuel_df.query(f"State == '{State}'").sort_values("ZIP", ascending=True).tail(valueRange).ZIP
    
    print(f"State: {State}\
    \n{State} Lower Numbered ZIP Codes: \n{ZIPCode_mins}\n\
    \n{State} Higher Numbered ZIP Codes: \n{ZIPCode_maxs}\n")
```

### Importing Population Dataset find the Most Populous States

#### Source

Since some of the Visualizations ask for Alternative Fuel Station data for the 10 Most Populous States, we got data from the 2020 Census Apportionment Results webpage on the US Census Bureau webstie. We downloaded an excel, file under `Table 2. Resident Population for the 50 States, the District of Columbia, and Puerto Rico: 2020 Census` heading, and saved it as a .csv file.

#### Reading in the Data

```python
census_df = pd.read_csv("apportionment-2020-table02.csv")
```

```python
census_df
```

#### Cleaning 

```python
#Dropping the Unnamed: 2 column. Original excel file had it left blank, so there is no need to keep it.
census_df.drop(columns = ['Unnamed: 2'], axis=1, inplace=True)

#Renaming the remaining two columns as State and Resident_Population accordingly
census_df.rename({"U.S. Department of Commerce":"State", "Unnamed: 1":"Resident_Population"}, axis=1, inplace=True)

#Dropping rows [0,1,2,54,56,57,58], since they do not pertain to single state names or single state populations
census_df.drop([0,1,2,54,56,57,58], axis=0, inplace=True)

#With dropping the rows in the previous line of code, the index will start at 3 with Alabama. 
#We drop decide to drop the index so that the index will start back at 0.
census_df.reset_index(drop=True,inplace=True)

#Checking cleaned dataframe
census_df
```

#### Sorting Data by Resident_Population to See Which States Have The Most Residents

```python
census_df.dtypes
```

**The Resident_Population values are stored as strings and also have commas between the numbers, but we need to remove the commas and convert the Residental_Population values to int values to sort and find the top 10**

```python
#census_df.Resident_Population.str.replace(",","") removes the commas from the Residential_Population values with no space
#so all that is left are the value of the population stored as a object.

census_df.Resident_Population = census_df.Resident_Population.str.replace(",","")

#Converting census_df.Resident_Population.dtype to int64, so the data can become numerical

census_df.Resident_Population = census_df.Resident_Population.astype('int64')
```

```python
#Sorting the state populations from highest to lowest and reseting the index so that the first index value is 0
census_df.sort_values(by="Resident_Population", ascending=False, inplace=True)
census_df.reset_index(drop=True, inplace=True)
census_df
```
